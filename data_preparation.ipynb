```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
names_labels_train = pd.read_csv("data/NR-ER-train/names_labels.csv", names=("name", "label"))
names_smiles_train = pd.read_csv("data/NR-ER-train/names_smiles.csv", names=("name", "smiles"))
data_train = names_smiles_train.join(names_labels_train.set_index("name"), on="name")

names_labels_test = pd.read_csv("data/NR-ER-test/names_labels.csv", names=("name", "label"))
names_smiles_test = pd.read_csv("data/NR-ER-test/names_smiles.csv", names=("name", "smiles"))
data_test = names_smiles_test.join(names_labels_test.set_index("name"), on="name")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), tight_layout=True)

sns.countplot(data=data_train, x="label", ax=ax1)
sns.countplot(data=data_test, x="label", ax=ax2)

ax1.set_xticks((0, 1), ("0 = Non Toxic", "1 = Toxic"))
ax2.set_xticks((0, 1), ("0 = Non Toxic", "1 = Toxic"))
ax1.set_title("Train")
ax2.set_title("Test")

plt.show()
```
